# **Data Loading**



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print('Module Installed Successfully!!!....')
```

```python
path = 'ecommerce_customer_data_quality_practice.csv'
df = pd.read_csv(path)
df.head()
```

```python
df['email'] = df['email'].fillna('Not Available')
df['gender'] = df['gender'].fillna('Not Available')
df['age'] = df['age'].fillna(df['age'].mean())
#df['customer_comments'] = df['customer_comments'].fillna('Not Available')
```

```python
df.head()
```

```python
df
```

```python
df['satisfaction_score'] = pd.to_numeric(df['satisfaction_score'], errors='coerce')
missing_percentage = df['satisfaction_score'].isnull().mean() * 100
mean_satisfaction = df['satisfaction_score'].mean()
df['satisfaction_score'] = df['satisfaction_score'].fillna(mean_satisfaction)
#print(f"Missing percentage after filling: {df['satisfaction_score'].isnull().mean() * 100:.2f}%")
```

```python
df['satisfaction_score'].isnull().mean()*100
```

```python
df.isnull().mean()*100
```

```python
df['income'] = pd.to_numeric(df['income'], errors='coerce')
df['income'] = df['income'].fillna(df['income'].mean())
df['income'].isnull().mean()*100
```

```python
df['state'] = df['state'].fillna('Not Available')
df['state'].isnull().mean()*100
```

```python
df['zip_code'] = df['zip_code'].fillna(000000)
df['zip_code'].isnull().mean()*100
```

```python
df['education_level'] = df['education_level'].fillna('Not Available')
df['education_level'].isnull().mean()*100
```

```python
df['favorite_category'] = df['favorite_category'].fillna('Not Available')
df['favorite_category'].isnull().mean()*100
```

```python
df['loyalty_program'] = df['loyalty_program'].fillna('Not Available')
df['loyalty_program'].isnull().mean()*100
```

```python
df['discount_usage_rate'] = pd.to_numeric(df['discount_usage_rate'], errors='coerce')
df['discount_usage_rate'] = df['discount_usage_rate'].fillna(df['discount_usage_rate'].mean())
df['discount_usage_rate'].isnull().mean()*100
```

```python
df['customer_comments'] = df['customer_comments'].replace('-', 'Not Available')
df['customer_comments'] = df['customer_comments'].fillna('Not Available')
df['customer_comments'].isnull().mean()*100
```

```python
df['gender'] = df['gender'].replace('F', 'Female')
df['gender'] = df['gender'].replace('M', 'Male')
df['gender'] = df['gender'].replace('mal', 'Male')
df['gender'] = df['gender'].replace('femal', 'Female')
df['gender'] = df['gender'].replace('-', 'Not Available-')
df['gender'].value_counts()
#
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='gender', data=df, palette='viridis', hue='gender', legend=False)
plt.title('Distribution of Customer gender')
plt.xlabel('gender')
plt.ylabel('Number of Customers')

# Add value on every bar with one decimal place
for container in plt.gca().containers:
    plt.bar_label(container)

plt.show()
```
